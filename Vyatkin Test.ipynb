```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dataset_Facebook.csv", delimiter=";")
```

```python
dfPhoto = df[df.Type == 'Photo']
dfStatus = df[df.Type == 'Status']
dfLink = df[df.Type == 'Link']
dfVideo = df[df.Type == 'Video']
```

```python
dftotal = df.describe().loc[["count","mean","min","max"]]
dfPhotoTotal = dfPhoto.describe().loc[["count","mean","min","max"]]
dfStatusTotal = dfStatus.describe().loc[["count","mean","min","max"]]
dfLinkTotal = dfLink.describe().loc[["count","mean","min","max"]]
dfVideoTotal = dfVideo.describe().loc[["count","mean","min","max"]]
```

```python
columns = list(dftotal.columns)
```

```python
Mode = pd.Series(name="mode")
dftotal = dftotal.append(Mode)
dfPhotoTotal = dfPhotoTotal.append(Mode)
dfStatusTotal = dfStatusTotal.append(Mode)
dfLinkTotal = dfLinkTotal.append(Mode)
dfVideoTotal = dfVideoTotal.append(Mode)
```

```python
for i in list(dftotal.columns):
    dftotal[i]["mode"] = np.array(df[i].mode(),object)
    dfPhotoTotal[i]["mode"] = np.array(dfPhoto[i].mode(),object)
    dfStatusTotal[i]["mode"] = np.array(dfStatus[i].mode(),object)
    dfLinkTotal[i]["mode"] = np.array(dfLink[i].mode(),object)
    dfVideoTotal[i]["mode"] = np.array(dfVideo[i].mode(),object)
```
